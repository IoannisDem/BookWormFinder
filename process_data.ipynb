```python
import os
from glob import glob
import cv2 
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

## Data

```python
data_path = "/home/pj00/projects/Github/BookWormFinder/data"

imgs = glob(os.path.join(data_path, "train/images/*.jpg"))
imgs = sorted(imgs, key = lambda x: x.split("/")[-1])

labels = glob(os.path.join(data_path, "train/labels/*.txt"))
labels = sorted(labels, key = lambda x: x.split("/")[-1])
```

## Understand data structure

```python
with open(labels[0]) as file:
    lines = [line.rstrip() for line in file]
```

```python
img = cv2.imread(imgs[0])
```

```python
plt.imshow(img)
plt.show()
```

```python
height, width, _ = img.shape
```

```python
class_index, x1, y1, x2, y2, x3, y3, x4, y4 = lines[0].split(" ")
class_index = int(class_index)
x1 = int(float(x1) * width)
y1 = int(float(y1) * height)
x2 = int(float(x2) * width)
y2 = int(float(y2) * height)
x3 = int(float(x3) * width)
y3 = int(float(y3) * height)
x4 = int(float(x4) * width)
y4 = int(float(y4) * height)
```

```python
cnt = np.array([
            [[x1, y1]],
            [[x2, y2]],
            [[x3, y3]],
            [[x4, y4]]
        ])
```

```python
rect = cv2.minAreaRect(cnt)
```

```python
box = cv2.boxPoints(rect)
box = box.astype(int)

print("bounding box: {}".format(box))
cv2.drawContours(img, [box], 0, (0, 0, 255), 2)
```

```python
plt.imshow(img)
plt.show()
```
